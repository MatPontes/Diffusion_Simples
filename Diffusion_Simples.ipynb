```python
import numpy as np
from PIL import Image
import cv2
import matplotlib.pyplot as plt
```

```python
# Função para simular a difusão de informação na imagem
def simulate_diffusion(image, original_image, iterations=10, diffusion_rate=0.1):
    for _ in range(iterations):
        # Aplicar filtro de desfoque gaussiano
        image = cv2.GaussianBlur(image, (3, 3), 0)
        image = image + diffusion_rate * (original_image - image)
    return image
```

```python
# Carregar uma imagem de exemplo
im_1 = np.array(Image.open('teste.jpg'))
```

```python
# Converter a imagem para escala de cinza
gray_image = cv2.cvtColor(im_1, cv2.COLOR_BGR2GRAY)
```

```python
# Adicionar ruído à imagem (simulação de degradação)
noisy_image = gray_image + np.random.normal(0, 25, gray_image.shape).astype(np.uint8)
```

```python
# Visualizar a imagem original e a imagem com ruído
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Imagem Original')

plt.subplot(1, 2, 2)
plt.imshow(noisy_image, cmap='gray')
plt.title('Imagem com Ruído')

plt.show()
```

```python
# Simular a difusão da informação na imagem com ruído (restauração)
restored_image = simulate_diffusion(noisy_image.copy(), gray_image)
```

```python
# Visualizar a imagem restaurada
plt.imshow(restored_image, cmap='gray')
plt.title('Imagem Restaurada')
plt.show()
```
